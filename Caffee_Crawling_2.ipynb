```python
import json

with open('probes.json','rt',encoding='utf-8') as file_object:
        # store file data in object
        data = json.load(file_object)
print(len(data))
```

```python
gus = []
for d in data:
    if '서울특별시' == d['name'][:5]:
        s = d['name'][6:].split(' ')
        gus.append(s[0])
```

```python
gus = set(gus[1:])
gus
```

```python
dic = {}
for gu in gus:
    dic[gu]=[]

for item in data:
    if item['name'][:5]=="서울특별시":
        s = item['name'][6:]
        gu = s.split(' ')[0]
        if gu in dic:
            dic[gu].append({'latitude':item['latitude'],'longitude':item['longitude'],'name':item['name']})
```

```python
import requests
```

```python
f = open('apk.txt')
apk = str(f.readline())
```

```python
cafes = {}
cafe_names = {}
for gu in gus:
    cafes[gu] = []
    cafe_names[gu] = []
    
for gu in dic:
    for pos in dic[gu]:
        lat = pos['latitude']
        long = pos['longitude']
        r = requests.get("https://dapi.kakao.com/v2/local/search/category.json?category_group_code=CE7&x="+str(long)+"&y="+str(lat)+"&radius=20000",headers={'Authorization':'KakaoAK '+apk})
        j_d = r.json()
        for place in j_d['documents']:
            if '카페' in place['category_name']:
                cafes[gu].append(place)
```

```python
sum_ = 0
for gu in cafes:
    print(gu,":", len(cafes[gu]))
    sum_ += len(cafes[gu])
print('total: ',sum_)
```
